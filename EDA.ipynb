### **Przegląd Zestawu Danych:**
Zestaw danych zawiera **10 atrybutów numerycznych** oraz **8 atrybutów kategorycznych**. Głównym celem analizy jest przewidywanie zmiennej **Revenue**, która wskazuje, czy odwiedzający dokonał zakupu podczas swojej sesji.

### **Szczegóły Atrybutów:**

#### **Atrybuty Numeryczne:**

1. **Administrative**  
   - Liczba stron związanych z funkcjami administracyjnymi, które użytkownik odwiedził w trakcie sesji.  
   - **Zakres:** 0–10
   - **Interpretacja:** Wskazuje, jak intensywnie użytkownik wchodził w interakcję z treściami administracyjnymi podczas sesji.

2. **Administrative Duration**  
   - Łączny czas (w sekundach) spędzony na stronach administracyjnych.  
   - **Zakres:** 0–2500 sekund
   - **Interpretacja:** Odzwierciedla zaangażowanie użytkownika w treści administracyjne. Dłuższy czas może wskazywać na większe zainteresowanie.

3. **Informational**  
   - Liczba stron informacyjnych odwiedzonych przez użytkownika.  
   - **Zakres:** 0–3
   - **Interpretacja:** Wskazuje na zainteresowanie użytkownika treściami informacyjnymi, takimi jak FAQ czy przewodniki.

4. **Informational Duration**  
   - Łączny czas (w sekundach) spędzony na stronach informacyjnych.  
   - **Zakres:** 0–2000 sekund
   - **Interpretacja:** Odzwierciedla zaangażowanie użytkownika w treści informacyjne. Dłuższy czas może wskazywać na większe zainteresowanie tymi treściami.

5. **Product Related**  
   - Liczba stron związanych z produktami odwiedzonych przez użytkownika.  
   - **Zakres:** 0–100
   - **Interpretacja:** Mierzy głębokość, z jaką użytkownik eksplorował oferty produktów. Wyższa wartość może wskazywać na większą chęć zakupu.

6. **Product Related Duration**  
   - Łączny czas (w sekundach) spędzony na stronach produktowych.  
   - **Zakres:** 0–5000 sekund
   - **Interpretacja:** Odzwierciedla czas spędzony na stronach produktów, co może sugerować poważniejsze rozważenie zakupu.

7. **Bounce Rate**  
   - Procent użytkowników, którzy weszli na stronę i natychmiast ją opuścili (tzw. "bounce").  
   - **Zakres:** 0–1 (0 do 100%)
   - **Interpretacja:** Wskazuje na skuteczność stron docelowych. Wyższy współczynnik odrzuceń oznacza, że użytkownicy nie angażują się w inne treści, co może sugerować niską jakość strony.

8. **Exit Rate**  
   - Procent użytkowników, którzy opuścili stronę, będąc jej ostatnim punktem w sesji.  
   - **Zakres:** 0–0.1 (0 do 10%)
   - **Interpretacja:** Pomaga zrozumieć, jak często ta strona jest ostatnią odwiedzaną w sesji. Wysoki współczynnik wyjść może sugerować, że użytkownicy zazwyczaj kończą swoją sesję na tej stronie.

9. **Page Value**  
   - Średnia wartość przypisana stronie, którą użytkownik odwiedził przed dokonaniem zakupu.  
   - **Zakres:** 0–100
   - **Interpretacja:** Odzwierciedla znaczenie strony w procesie konwersji. Strony z wyższymi wartościami są tymi, które najczęściej poprzedzają transakcje.

10. **Special Day**  
    - Wartość wskazująca na bliskość sesji do konkretnego dnia specjalnego (np. Dzień Matki, Walentynki).  
    - **Zakres:** 0–1
    - **Interpretacja:** Określa, jak blisko daty sesji znajduje się specjalny dzień, który może wpłynąć na skłonność użytkownika do zakupu. Wartość ta jest wyższa w dniach przed i po świętach, gdy transakcje są bardziej prawdopodobne.

#### **Atrybuty Kategoryczne:**

1. **Month**  
   - Miesiąc, w którym miała miejsce wizyta.  
   - **Możliwe wartości:** "Jan", "Feb", "Mar", ..., "Dec"
   - **Interpretacja:** Może pomóc w analizie sezonowych trendów i wpływu poszczególnych miesięcy na generowanie przychodu.

2. **Operating Systems**  
   - Typ systemu operacyjnego używanego przez odwiedzającego (np. Windows, macOS).  
   - **Zakres:** 1–3 (każda liczba odpowiada innemu systemowi operacyjnemu)
   - **Interpretacja:** Pokazuje, które systemy operacyjne są najczęściej używane przez odwiedzających.

3. **Browser**  
   - Typ przeglądarki internetowej, z której korzysta użytkownik (np. Chrome, Firefox, Safari).  
   - **Zakres:** 1–5 (każda liczba odpowiada innej przeglądarce)
   - **Interpretacja:** Umożliwia analizę, które przeglądarki są najczęściej używane przez odwiedzających, co może pomóc w optymalizacji strony.

4. **Region**  
   - Region geograficzny użytkownika.  
   - **Zakres:** 1–5 (każda liczba odpowiada innemu regionowi)
   - **Interpretacja:** Pomaga zrozumieć różnice regionalne w zachowaniach użytkowników i wydajności strony, co może być pomocne w lokalizacji i marketingu.

5. **Traffic Type**  
   - Źródło ruchu użytkownika (np. bezpośredni, polecenie, płatna reklama).  
   - **Zakres:** 1–3 (każda liczba odpowiada innemu typowi ruchu)
   - **Interpretacja:** Informacje te są istotne dla analizy skuteczności różnych źródeł ruchu i ich konwersji.

6. **Visitor Type**  
   - Określa, czy użytkownik jest nowym odwiedzającym, czy powracającym.  
   - **Możliwe wartości:** "Returning_Visitor", "New_Visitor"
   - **Interpretacja:** Pomaga rozróżnić odwiedzających, którzy już wcześniej byli na stronie, od nowych, którzy mogą wymagać większej zachęty do dokonania zakupu.

7. **Weekend**  
   - Określa, czy wizyta miała miejsce w weekend.  
   - **Możliwe wartości:** `TRUE`, `FALSE`
   - **Interpretacja:** Wizyty weekendowe mogą różnić się pod względem zachowań użytkowników, ponieważ niektórzy mogą częściej dokonywać zakupów w weekendy.

8. **Revenue**  
   - Zmienna docelowa, wskazująca, czy użytkownik dokonał zakupu.  
   - **Możliwe wartości:** `TRUE`, `FALSE`
   - **Interpretacja:** To wynik, który chcemy przewidzieć — czy użytkownik dokona zakupu podczas swojej sesji.

<div class="alert alert-primary" role="alert">
  Sprawdzenie multikolinearności (VIF)

Variance Inflation Factor
-spradzanie współliniowosści między zmiennymi niezależymi 

| VIF   | Interpretacja                                    |
| ----- | ------------------------------------------------ |
| = 1   | Brak współliniowości                             |
| 1 – 5 | Akceptowalna współliniowość                      |
| > 5   | Potencjalny problem  |
| > 10  | Silna współliniowość           |


</div>

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from main import walidacja

X_train, X_test, y_train,y_test = walidacja()

X_vif = add_constant(X_train)

vif_data = pd.DataFrame()

vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print(vif_data)

```

```python
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import StandardScaler

wrong_data = []
for key,value in vif_data.values:
    if key != 'const'and value > 4.5:
        wrong_data.append(key)

X_train_reduced = X_train.drop(columns=wrong_data)
X_test_reduced = X_test.drop(columns=wrong_data)

scaler = StandardScaler()
X_train_reduced_Scaled = scaler.fit_transform(X_train_reduced)
X_test_reduced_Scaled = scaler.transform(X_test_reduced)

# Regularyzcja danych "penalty=l2" ridge
model_ridge_weighted = LogisticRegression(penalty='l2', solver='saga', max_iter=10000, class_weight='balanced')

model_ridge_weighted.fit(X_train_reduced_Scaled, y_train)

y_pred_ridge_weighted = model_ridge_weighted.predict(X_test_reduced_Scaled)
y_proba_ridge_weighted = model_ridge_weighted.predict_proba(X_test_reduced_Scaled)[:, 1]

print("Ridge (z ważeniem klas) - AUC:", roc_auc_score(y_test, y_proba_ridge_weighted))
print("Ridge (z ważeniem klas) - Classification Report:")
print(classification_report(y_test, y_pred_ridge_weighted))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support

thresholds = np.arange(0.0, 1.05, 0.05)
f1_scores = []
precisions = []
recalls = []

for t in thresholds:
    preds = (y_proba_ridge_weighted >= t).astype(int)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, preds, average='binary')
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

plt.figure(figsize=(10,6))
plt.plot(thresholds, f1_scores, label='F1-score')
plt.plot(thresholds, precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision, Recall, F1 vs Threshold')
plt.legend()
plt.grid(True)
plt.show()


precisions = np.array(precisions)
recalls = np.array(recalls)
diff = np.abs(precisions - recalls)
min_idx = np.argmin(diff)

best_threshold = thresholds[min_idx]
best_precision = precisions[min_idx]
best_recall = recalls[min_idx]

print(f"Punkt przecięcia: threshold = {best_threshold:.2f}, precision = {best_precision:.2f}, recall = {best_recall:.2f}")
```

```python
threshold = 0.59
y_pred_custom_thresh = (y_proba_ridge_weighted >= threshold).astype(int)

print(f"Classification Report (próg={threshold}):")
print(classification_report(y_test, y_pred_custom_thresh))

from sklearn.metrics import confusion_matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_custom_thresh))
```
